```python
prompt = """
당신은 피자가게 손님입니다. 유명한 피자가게의 건너편에 새로운 피자가게가 생겨 오늘 처음 방문하였습니다.
당신은 새로운 피자가게가 유명한 피자가게를 이길 수 있는지 까다롭게 평가하고 싶습니다.
까다롭고 제대로 이해하기 힘든 주문으로 새로운 피자가게의 피자를 주문해보세요.

### 지시 사항 ###
원하는 피자의 토핑을 주문해야 합니다. 
토마토 소스와 치즈는 기본 토핑입니다.
추가 피자 토핑의 종류는 페퍼로니, 소시지, 피망, 가지가 있습니다.
토핑은 1 ~ 2개 들어갈 수 있습니다.
하나의 피자만 주문 하세요.
사용해야하는 토핑도 같이 출력해주세요.

### 주문 예시 ### 
# 치즈 피자 
{"제 친구는 채식주의자예요. 근데 야채랑 과일을 안 좋아해요. 버섯류는 좋아하는데, 피자에는 별로래요.": ['도우', '소스', '치즈']}
{"치즈 하나요.": ['도우', '소스', '치즈']}
{"아침의 태양처럼 밝고 아름다운 피자": ['도우', '소스', '치즈']}
{"페퍼로니 없는 페퍼로니 피자": ['도우', '소스', '치즈']}

# 페퍼로니 피자
{"프로볼로니": ['도우', '소스', '치즈', '페퍼로니']}
{"마카로니": ['도우', '소스', '치즈', '페퍼로니']}
{"짜요": ['도우', '소스', '치즈', '페퍼로니']}
{"살라미": ['도우', '소스', '치즈', '페퍼로니']}
{"요자피 니로퍼페": ['도우', '소스', '치즈', '페퍼로니']}
{"ㅍㅍㄹㄴ ㅍㅈ ㅎㄴㅇ": ['도우', '소스', '치즈', '페퍼로니']}
   
# 소시지 피자
{"고기 들어간 거요": ['도우', '소스', '치즈', '소시지']}
{"ㅅㅅㅈ ㅍㅈ": ['도우', '소스', '치즈', '소시지']}
    
# 피망 피자
{"ㅍㅁ ㅍㅈ ㅎㄴㅇ": ['도우', '소스', '치즈', '피망']}
{"피망과 소시지요. 아, 소시지 빼주세요!": ['도우', '소스', '치즈', '피망']}
{"ㅍㅁ피자": ['도우', '소스', '치즈', '피망']}
{"초록색 야채 들어간 거": ['도우', '소스', '치즈', '피망']}
    
# 가지 피자
{"가지가지한 피자": ['도우', '소스', '치즈', '가지']}
{"보라색 토마토 비슷한 거": ['도우', '소스', '치즈', '가지']}
{"음, 어떤 피자를 먹어야 잘 먹었다고 소문이 가~지?": ['도우', '소스', '치즈', '가지']}
    
# 두 가지 토핑 피자
{"페퍼로니랑 소시지랑 사귀는 거 맞죠?": ['도우', '소스', '치즈', '페퍼로니', '소시지']}
{"저는 페퍼로니와 피망이 같이 구워졌을 때의 향을 좋아해요.": ['도우', '소스', '치즈', '페퍼로니', '피망']}
{"소시지랑 피망 둘 다 주세요": ['도우', '소스', '치즈', '소시지', '피망']}
{"가지랑 피망이요": ['도우', '소스', '치즈', '가지', '피망']}

## 출력 형태 ##
{"주문 내용":[사용한 토핑]}
"""
```

```python
import ollama

def customer_order(prompt):
    response = ollama.chat(
        model='lancard/korean-yanolja-eeve',
        messages=[
            {"role": "system", "content": f"{prompt} 응답은 100자 이내로 작성해 주세요."}, 
            {"role": "user", "content": "까다롭고 제대로 이해하기 힘든 주문을 만들어 주세요."}
        ]
    )
    return response['message']['content']  
```

```python
response = customer_order(prompt)
print(response)
```

```python
import re

toppings = re.search(r'사용된 토핑:\s*(.*)', response)
toppings = toppings.group(1).split(', ') if toppings else []

order = response.split('사용된 토핑:')[0].strip()

# 출력
print("주문:", order)
print("사용된 토핑:", toppings)
```

```python
# 재료비 설정
ingredients_cost = {
    '도우': 2000, 
    '소스': 600, 
    '치즈': 600,  
    '페퍼로니': 2880,  
    '소시지': 2880,  
    '피망': 1620,  
    '가지': 1260,  
}

# 가격 설정
pizza_base_price = {
    '도우': 5000, 
    '소스': 1000, 
    '치즈': 1000,  
    '페퍼로니': 3000,  
    '소시지': 3000,  
    '피망': 1800,  
    '가지': 1300,  
}

def calculate_profit(ingredients):
    # 재료비 계산
    total_cost = 0
    for ingredient in ingredients:
        if ingredient in ingredients_cost:
            total_cost += ingredients_cost[ingredient]
    
    # 판매 가격 계산 
    total_price = 0
    for ingredient in ingredients:
        if ingredient in pizza_base_price:
            total_price += pizza_base_price[ingredient]
    
    # 이익 계산
    profit = total_price - total_cost
    return profit
```

```python
print(calculate_profit(toppings))
```
